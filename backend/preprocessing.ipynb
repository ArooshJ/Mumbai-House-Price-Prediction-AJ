```python
# -- used ChatGPT for syntax + personal tweaks ---
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="geoapi")

def get_lat_long(location):
    print(f"fetching {location}")
    try:
        loc = geolocator.geocode(location + ", Mumbai")
        return loc.latitude, loc.longitude
    except:
        print(f"Error in fetching location {location}, searching MMR")
        try:
            loc = geolocator.geocode(location + ", Mumbai Metropolitan Region")
            return loc.latitude, loc.longitude
        except:
            print(f"Not even found in MMR") 
            try:
              loc = geolocator.geocode(location + ", Navi Mumbai")
              print(f"F")
              return loc.latitude, loc.longitude
            except:
              try:
                loc = geolocator.geocode(location + ", District Thane")
                print(f"Found in Thane")
                return loc.latitude, loc.longitude
              except:
                 print("Not found")  
               
            
    return None, None
```

```python

lat_long_map = {}

for loc in df['Location'].unique():
    lat_long_map[loc] = get_lat_long(loc)
    print(f"latlong map [loc] = {lat_long_map[loc]}")
```

```python
null_locs = [key for key, value in lat_long_map.items() if value[0] is None]
null_locs
```

Mapping various sectors and sub-areas with no lat_long to their main areas' lat_long

```python
for sub_region in null_locs:
    for parent_region in lat_long_map.keys():
        if parent_region in sub_region and parent_region != sub_region:  # Check if parent region is a substring
            lat_long_map[sub_region] = lat_long_map[parent_region]
            print(f"Mapping {sub_region} to {parent_region}: {lat_long_map[sub_region]}")
            null_locs.remove(sub_region)
            break  # Stop checking once we've found a match
```

```python
len(lat_long_map)
lat_long_map['Thane']
```

```python
def get_lat_long(location):
    if location in lat_long_map:
        return lat_long_map[location]
    else:
        # Fallback to check for substring matches in the lat_long_map
        for parent_location in lat_long_map.keys():
            if parent_location in location:
                return lat_long_map[parent_location]
    return None  # Return None if not found

# Applying the function to create new columns for latitude and longitude
df['Coordinates'] = df['Location'].apply(get_lat_long)
df[['Latitude', 'Longitude']] = pd.DataFrame(df['Coordinates'].tolist(), index=df.index)

# Drop the 'Coordinates' column if no longer needed
df.drop(columns=['Coordinates'], inplace=True)
```
